```python
from numpy import *
import matplotlib.pyplot as plt
import pylab

m = 5.97*10**24     #m - Massa do planeta
M = 1.99*10**30     #M - Massa do sol
a = 1.49*10**8      #a - Semi-eixo maior
G = 6.674*10**-11   #G - Constante gravitacional        
E = -G*M*m/(2*a)   #E - Energia
L = 9.11*10**38     #L - Momento angular
p = L**2/(m*G*M*m)  #p - Semi-latus
c = 1 + (2*E*L**2)/(m*(G*M*m)**2)

if c<=0:
  e = sqrt(-c)        #e = Excentricidade
else:
  e = sqrt(c)


def fx(x):
    r = p/(1 + e*cos(x))
    return r

n = 1000
phi = linspace(0,2*pi,n)
radius = zeros([n])
theta = zeros([n])
x = zeros([n])
y = zeros([n])

for i in range(0,n):
    radius[i] = fx(phi[i])
    theta[i] = 180*phi[i]/pi

for i in range(0,n):
    x[i] = radius[i]*cos(phi[i])

for i in range(0,n):
    y[i] = radius[i]*sin(phi[i])

pylab.xlabel("x (Km)")
pylab.ylabel("y (Km)")
pylab.plot(x,y, label = "Orbita")

pylab.scatter(0,0, s = 200, c = 'yellow', alpha = 1)

pylab.legend()
pylab.grid()
pylab.show()
```
